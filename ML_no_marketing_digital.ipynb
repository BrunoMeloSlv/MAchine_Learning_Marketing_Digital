```python
import pandas as pd

dados = pd.read_csv('https://raw.githubusercontent.com/alura-cursos/digital-marketing-ml/master/train.csv', dtype={
    'date':object,
    'fullVisitorId':object,
    'visitId':object
})
print(dados.shape)
dados.head()
```

```python
len(dados.fullVisitorId.unique())
```

```python
dados.dtypes
```

```python
import json

json.loads(dados.device[0])
```

```python
pd.DataFrame([json.loads(linha) for linha in dados.device])
```

```python
dicionario = ['device','geoNetwork','trafficSource','totals']
```

```python
for coluna in dicionario:
    dados = dados.join(pd.DataFrame([json.loads(linha) for linha in dados[coluna]]))

print(dados.shape)
dados.head()
```

```python
dados.drop(dicionario, axis=1, inplace=True)
dados.head()
```

```python
dados.drop('adwordsClickInfo', axis=1, inplace= True)
```

```python
for coluna in dados.columns:
    print(coluna + ': ' + str(len(dados[coluna].unique())))
```

```python
coluna_na = []
for coluna in dados.columns:
    if len(dados[coluna].unique()) == 1:
        coluna_na.append(coluna)

dados.drop(coluna_na, axis=1, inplace= True)
dados.head()
```

```python
dados.dtypes
```

```python
quant = ['bounces','hits','newVisits','pageviews','transactionRevenue']

for coluna in quant:
    dados[coluna] = pd.to_numeric(dados[coluna])

dados.dtypes
```

```python
dados.head()
```

```python
dados['transactionRevenue'].isna().sum()
```

```python
dados['transactionRevenue'].fillna(0, inplace=True)
```

```python
dados['transactionRevenue'] = dados['transactionRevenue']/1000000
dados
```

```python
df_quant = dados.groupby('fullVisitorId', as_index= False )[quant].sum()
df_quant
```

```python
y = df_quant['transactionRevenue'].copy()
x = df_quant.drop('transactionRevenue', axis=1)
```

```python
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression

x_treino, x_teste, y_treino, y_teste = train_test_split(x,y, test_size = 0.3, random_state= 42)
reg = LinearRegression()
reg.fit(x_treino,y_treino)
reg_predict = reg.predict(x_teste)
```

```python
from sklearn.metrics import mean_squared_error
import numpy as np

np.sqrt(mean_squared_error(y_teste, reg_predict))
```

##### Retornando a última visita

```python
visitas_ultima = dados.groupby('fullVisitorId', as_index= False)
visitas_ultima = visitas_ultima['visitNumber'].max()  
```

```python
usuarios_visitas_unicas = dados.drop_duplicates(subset=['fullVisitorId','visitNumber'])
usuarios_visitas_unicas
```

```python
visitas = pd.merge(visitas_ultima,usuarios_visitas_unicas, left_on=['fullVisitorId','visitNumber'],
                   right_on=['fullVisitorId','visitNumber'], how='left')
visitas
```

##### Retornando a primeira visita

```python
visitas_primeira = dados.groupby('fullVisitorId', as_index= False)
visitas_primeira = visitas_primeira['visitNumber'].min()  
```

```python
visitas_primeira.set_index('fullVisitorId', inplace= True)
```

```python
visitas = visitas.join(visitas_primeira, how= 'left', on=['fullVisitorId'], rsuffix='primeira')
```

```python
visitas
```

```python
visitas = pd.merge(visitas, usuarios_visitas_unicas, left_on=['fullVisitorId','visitNumberprimeira'],
                   right_on=['fullVisitorId','visitNumber'], how='left', suffixes=['_ultima','_primeira'])
```

##### Limpando a Base

```python
for coluna in quant:
    visitas.drop(coluna + '_ultima'   , axis= 1, inplace= True)
    visitas.drop(coluna + '_primeira' , axis= 1, inplace= True)
```

```python
ids = ['sessionId_ultima','visitId_ultima','sessionId_primeira','visitId_primeira']
visitas.drop(ids, axis= 1, inplace= True)
```

```python
visitas.columns
```

```python
geo = ['city_primeira','continent_primeira','subContinent_primeira','metro_primeira','country_primeira',
       'region_primeira','networkDomain_primeira']

visitas.drop(geo, axis= 1, inplace= True)
```

```python
visitas
```

##### Criando Novas Variáveis

```python
visitas = pd.merge(visitas, df_quant, left_on=['fullVisitorId'],
                   right_on=['fullVisitorId'], how='left')
```

```python
visitas['tempo_dif'] = visitas.visitStartTime_ultima - visitas['visitStartTime_primeira']
```

```python
visits = dados.groupby('fullVisitorId', as_index=False).count()['visitNumber'].values
```

```python
visitas['visits'] = visits
```

```python
visitas['ano_ultima'] = pd.to_numeric([data[0:4] for data in visitas.date_ultima])
visitas['mes_ultima'] = pd.to_numeric([data[4:6] for data in visitas.date_ultima])
visitas['dia_ultima'] = pd.to_numeric([data[6:8] for data in visitas.date_ultima])

visitas['ano_primeira'] = pd.to_numeric([data[0:4] for data in visitas.date_primeira])
visitas['mes_primeira'] = pd.to_numeric([data[4:6] for data in visitas.date_primeira])
visitas['dia_primeira'] = pd.to_numeric([data[6:8] for data in visitas.date_primeira])
```

```python
visitas
```

```python
visitas.drop('fullVisitorId',axis= 1, inplace= True)
```

```python
y = visitas['transactionRevenue'].copy()
x = visitas.drop('transactionRevenue', axis=1)
```

```python
x.head()
```

```python
quali = visitas.dtypes[visitas.dtypes == object].keys()
```

```python
from sklearn.preprocessing import LabelEncoder

for coluna in quali:
    lbl = LabelEncoder()
    strings = list(x[coluna].values.astype('str'))
    lbl.fit(strings)
    x[coluna] = lbl.transform(strings)
```

```python
x.head()
```

```python
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression

x_treino, x_teste, y_treino, y_teste = train_test_split(x,y, test_size = 0.3, random_state= 42)
reg = LinearRegression()
reg.fit(x_treino,y_treino)
reg_predict = reg.predict(x_teste)
```
